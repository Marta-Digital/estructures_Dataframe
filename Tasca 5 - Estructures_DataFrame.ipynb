__3. Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).__

__-Añadir una columna con la velocidad media de vuelo.__

```python
file['V_Media']= ((file['Distance'].mean()/file['CRSElapsedTime'].mean())*60).round(2)
file[:3]
```

__-Si el avión ha llegado con retraso o no.__ 

```python
file['Con_retraso']=file.ArrDelay>=0
file[:5]
```

__-Añadir una columna con la velocidad de cada vuelo.__

```python
file['Velocidad']= (file['Distance']/file['CRSElapsedTime']*60).round(2)
file[:3]
```

__-Añadir una columna fecha, convertirla en el índice__

```python
fechas = file['Year'].astype(str) + "-" + file['Month'].astype(str) + "-" + file["DayofMonth"].astype(str)
fechas [:5]
```

```python
file["Date"] = pd.to_datetime(fechas)
file.Date[:5]
```

```python
file.set_index("Date", inplace=True)
file[:5]
```

__4. Taula de les aerolínies amb més endarreriments acumulats.__

```python
file["TotalDelay"]=file['CarrierDelay']+file['WeatherDelay']+file['NASDelay']+file['SecurityDelay']+file['LateAircraftDelay']
file[:3]
```

```python
delay = file[['UniqueCarrier','TotalDelay']].copy()
delay_ok=delay.dropna().sort_values("TotalDelay", ascending=False)
delay_ok[:5]
```

```python
delay_ok=delay_ok.groupby('UniqueCarrier').aggregate(sum)
delay_ok.sort_values("TotalDelay", ascending=False)

#La compañía WN es la que acumula con diferencia más retrasos, con un total de 11645 minutos (187.75 horas)
```

__7. ¿Cuales son los 5 aeropuertos que han tenido más retraso en la salida?.__

```python
AirDelays=file[['DepDelay','Origin']].copy()
```

```python
AirportDelays= AirDelays.groupby('Origin').aggregate(sum)
AirportDelays.sort_values('DepDelay', ascending=False)[:5]
```

__7. ¿Cuales son los que han recorrido más kilómetros a lo largo de 2018?__

```python
LongDist=pd.DataFrame(file, columns=["Distance","FlightCode"]).copy()
LongDist.set_index("FlightCode", inplace=True)
LongDist.sort_values('Distance', ascending=False).head()
```

```python
LongDist_ok=LongDist.groupby('FlightCode').aggregate(sum)
LongDist_ok
LongDist_sort=LongDist_ok.sort_values('Distance', ascending=False)
LongDist_sort[:5] 
#Los vuelos CO-15, AA-267 y DL-1273 son los 3 vuelos que recorren más Km.
```
